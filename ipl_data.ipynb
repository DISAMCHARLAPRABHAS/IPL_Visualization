```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def train_model(filepath):
  df=pd.read_csv(filepath)
  df.drop(columns=["umpire3"],inplace=True)
  df.dropna(subset=['umpire1', 'umpire2', 'player_of_match'], inplace=True)
  df['city'].fillna(df['city'].mode()[0], inplace=True)
  return df
```

```python
df=train_model('matches.csv')
df.head()
```

# Here the fetaures are input and target is output based on this we are going predict the data

```python
features = ['season', 'city', 'team1', 'team2', 'toss_winner', 'toss_decision', 'venue', "winner"]
target = 'win_by_runs'
X_train= df[features]
y_train = df[target]
```

```python
y_mean=y_train.mean().round(2)
y_pred_baseline=[y_mean]*len(y_train)
mae_baseline=mean_absolute_error(y_train, y_pred_baseline)
print("mean of targe", y_mean)
print("Mean Absolute Error (Baseline):", mae_baseline)
```

# we can use simpleimputer to handle the missing data in each column

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
X_train=imputer.fit_transform(X_train)
```

# one hot encoder using for change categorical data to numerical data

```python
from category_encoders import OneHotEncoder
ohe=OneHotEncoder(use_cat_names=True, handle_unknown='ignore')
XT_train=ohe.fit_transform(X_train)

print(XT_train.shape)
XT_train.head()
```

```python
from sklearn.linear_model import LinearRegression, Ridge  # noqa F401
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
```

```python
model=make_pipeline(
    OneHotEncoder(use_cat_names=True, handle_unknown='ignore'),
    Ridge()
)
model.fit(X_train, y_train)
```

```python
y_pred_training=model.predict(X_train)
mae=mean_absolute_error(y_train, y_pred_training)
print("Mean Absolute Error:", mae)
```

```python
intercept = model.named_steps["ridge"].intercept_
coefficients =  model.named_steps["ridge"].coef_
print("coefficients len:", len(coefficients))
print(coefficients[:5])  # First five coefficients
```

```python
feature_names = model.named_steps["onehotencoder"].get_feature_names()
print("features len:", len(feature_names))
print(feature_names[:5])  # First five feature names
```

```python
feat_imp = pd.Series(coefficients, feature_names)
feat_imp.head()
#feat_imp.shape
```

```python
feat_imp.sort_values(key=abs).tail(15).plot(kind="barh")
plt.xlabel("Regression Coefficient (Impact on Win by Runs)")
plt.ylabel("Encoded Feature (Team / Venue / Category)")
plt.title("Effect of Match Features on Win Margin (Runs) — Linear Regression Coefficients");
```

This graph shows how different teams, venues, and match contexts influence the margin by which a team wins (in runs).

# **Key** **Insights**:
3_Gujarat Lions has the strongest positive effect
→ When Gujarat Lions plays (likely as team1), the model predicts a larger win by runs.
This team contributes positively to dominating wins.


7_Gujarat Lions has the strongest negative effect
→ When this feature appears (possibly in another role), the win margin drops significantly, even negative.
 This indicates that matches involving them lead to closer games or losses.

Stadiums like Maharashtra Cricket Association Stadium and Nehru Stadium
→ Slightly increase win margins
 These venues may favor teams that win by runs.

Some Rising Pune Supergiant combinations (like 2_, 4_)
→ Have mixed or negative influence, possibly not strong performers in terms of run victories.

Short bars near 0
→ These categories have minimal effect on the win margin, and may be statistically insignificant.

# **Conclusion**:
The model has identified that certain teams and venues significantly impact how much a team wins by in terms of runs.
3_Gujarat Lions leads to stronger wins, while 7_Gujarat Lions and some Rising Pune combinations reduce the win-by-runs margin.
