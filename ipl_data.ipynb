```python
# prompt: make a dashboard with all this eda
# with in single sheet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('matches.csv')

# Display basic information
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
df.info()
print("\nFirst 10 rows of the dataset:")
print(df.head(10))

# Display top winners
top_winners = df["winner"].value_counts().head()
print("\nTop 5 winning teams:")
print(top_winners)

# Display average win metrics
avg_runs = df['win_by_runs'].mean()
avg_wckts = df['win_by_wickets'].mean().astype(int)
print("\nAverage win by runs:", avg_runs)
print("Average win by wickets:", avg_wckts)

# Display most frequent toss winner and match winner
toss_mode = df["toss_winner"].mode()
match_mode = df["winner"].mode()
print("\nMost frequent toss winner:", toss_mode[0])
print("Most frequent match winner:", match_mode[0]) # Using [0] to get the value from the Series

# Display descriptive statistics for win margins
print("\nDescriptive statistics for win margins:")
print(df[['win_by_runs', 'win_by_wickets']].describe())

# Set plot style
plt.style.use('ggplot')
sns.set_palette('Set2')

# --- Plotting Section ---

# Histogram of Win by Runs
plt.figure(figsize=(8, 4))
plt.hist(df["win_by_runs"], bins=20, rwidth=0.9, color='skyblue', edgecolor='black')
plt.title('Histogram of Win by Runs')
plt.xlabel('Win by Runs')
plt.ylabel('Frequency')
plt.grid(axis="y", alpha=0.75)
plt.show() # Display this plot

# Boxplot of Win by Runs
plt.figure(figsize=(8, 4))
plt.boxplot(df['win_by_runs'], vert=False)
plt.title('Boxplot of Win by Runs')
# plt.ylabel('Win by Runs') # This label is not meaningful for a horizontal boxplot
plt.xlabel('Win by Runs')
plt.show() # Display this plot

# Bar chart of Matches Won by Each Team
plt.figure(figsize=(10, 6)) # Adjusted figure size for better readability
win_teams = df["winner"].value_counts()
win_teams.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Matches Won')
plt.xticks(rotation=90) # Rotate labels for better readability
plt.grid(axis="y", alpha=0.75)
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show() # Display this plot

# Bar chart of Tosses Won by Each Team
plt.figure(figsize=(10, 6)) # Adjusted figure size
toss = df["toss_winner"].value_counts()
toss.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Number of Tosses Won by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Tosses Won')
plt.xticks(rotation=90) # Rotate labels
plt.grid(axis="y", alpha=0.75)
plt.tight_layout()
plt.show() # Display this plot

# Pie chart of Toss Decision
plt.figure(figsize=(6, 4))
toss_win = df["toss_decision"].value_counts()
toss_win.plot(kind='pie', autopct='%1.1f%%', explode=(0, 0.1), colors=['lightblue', 'lightgreen'])
plt.title('Toss Decision Distribution')
plt.ylabel('') # Remove default ylabel
plt.show() # Display this plot

# Pie chart of Match Result Distribution
plt.figure(figsize=(6, 4))
result_counts = df['result'].value_counts()
result_counts.plot(kind="pie", colors=['lightblue', 'lightgreen', 'lightcoral'], autopct='%1.1f%%', explode=(0, 0.3, 0.3))
plt.title('Match Result Distribution')
plt.ylabel('') # Remove default ylabel
plt.show() # Display this plot

# Bar chart of Top 10 Players of the Match
plt.figure(figsize=(10, 6))
potm = df["player_of_match"].value_counts().head(10).sort_values(ascending=False)
potm.plot(kind="bar", color="lightskyblue", edgecolor="black")
plt.title("Top 10 Player of the Match Awards")
plt.xlabel("Player")
plt.ylabel("Number of Awards")
plt.xticks(rotation=45, ha='right') # Adjust rotation and alignment
plt.tight_layout()
plt.show() # Display this plot

# Bar chart of Average Win by Runs by each Winning Team
plt.figure(figsize=(10, 6))
(
     df.groupby("winner")["win_by_runs"].mean()
    .sort_values(ascending=False)
    .plot(
        kind="bar",
          color="lightskyblue",
          edgecolor="black",
        xlabel="Team",
        ylabel="Average Win by Runs",
        title="Average Win by Runs by each Winning Team"
          )
);
plt.xticks(rotation=90) # Rotate labels
plt.tight_layout()
plt.show() # Display this plot

# Bar chart of Average Win by Wickets by each Winning Team
plt.figure(figsize=(10, 6))
(
     df.groupby("winner")["win_by_wickets"].mean()
    .sort_values(ascending=False)
    .plot(
        kind="bar",
          color="lightskyblue",
          edgecolor="black",
        xlabel="Team",
        ylabel="Average Win by wickets",
        title="Average Win by wickets by each Winning Team"
          )
);
plt.xticks(rotation=90) # Rotate labels
plt.tight_layout()
plt.show() # Display this plot

# Bar chart of Number of matches held in cities
plt.figure(figsize=(10, 6))
city = df["city"].value_counts()
city.plot(kind="bar", color="lightgreen", edgecolor="black", xlabel="city", ylabel="no of matches", title="Number of matches held in cities")
plt.xticks(rotation=90) # Rotate labels
plt.tight_layout()
plt.show() # Display this plot

# Line chart of Matches Played by Season
plt.figure(figsize=(8, 5))
df["season"].value_counts().sort_index().plot(kind="line", marker="o", color="lightgreen")
plt.title('Matches Played by Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(df['season'].sort_values().unique()) # Ensure all seasons are shown
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show() # Display this plot
```
