```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data

df = pd.read_csv('match_info_data.csv')

# Group by 'winner' to count total wins for each team
team_wins = df['winner'].value_counts()

# Plot a bar chart
team_wins.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Add labels and title
plt.title('Number of Wins by Each Team', fontsize=16)
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'C:/Users/Admin/match_info_data.csv'
df = pd.read_csv(file_path)

# Calculate the proportion of toss decisions
toss_decisions = df['toss_decision'].value_counts()

# Plot the pie chart
toss_decisions.plot(kind='pie', autopct='%1.1f%%', explode=(0,0.1), figsize=(6,4), colors=['lightblue', 'lightgreen'])

# Add a title
plt.title('Proportion of Toss Decisions', fontsize=16)
sns.set_style('darkgrid')
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'C:/Users/Admin/match_info_data.csv'
df = pd.read_csv(file_path)

# Check the 'result' column
print(df['result'].value_counts())

# Count the occurrences of each result type
result_counts = df['result'].value_counts()

# Prepare labels and sizes for the pie chart
labels = result_counts.index  # Result types
sizes = result_counts.values  # Counts

# Plot a pie chart
colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', 'lightcoral', 'lightskyblue' , 'gold']
explode=(0, 0.3)
plt.figure(figsize=(6,4))
plt.pie(sizes, explode=explode, labels=labels,  autopct='%1.1f%%',startangle=140, colors=colors)
plt.title('Percentage of Result Methods')
plt.axis('equal')
plt.legend(labels=result_counts.index, loc="best")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'C:/Users/Admin/match_info_data.csv'
df = pd.read_csv(file_path)

# Plot a histogram for win_by_runs
df['win_by_runs'].plot(kind='hist', bins=12, color='skyblue', edgecolor='black', figsize=(10, 6))

# Add labels and title
plt.title('Distribution of Wins by Runs', fontsize=16)
plt.xlabel('Win Margin (Runs)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
